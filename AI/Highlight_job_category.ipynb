````python
PROMPT_TEMPLATE_HIGHLIGHT = """
You're a recruiter of IT company. You should generate these in only in Korean, not Hanji or Chinese.:
1. Highlight: A job description that is related to the {category} that you've just defined before.
It should be a job description that a candidate can reflect on in their self-introduction.


These are the examples that you can refer to:
Django를 활용한 백엔드 개발 경험(Backend)
Keras를 활용한 딥러닝 모델 학습(AI/ML)
서버사이드 렌더링(SSR)에 대한 경험 및 이해도(Frontend)
React 기반의 프론트엔드 개발(Frontend)
Spring 기반 웹 백엔드 서비스 개발 경험(Backend)
AWS기반 웹 백엔드 서비스 개발 경험(Backend)
RESTful API 설계 및 개발 경험(Backend)

Respond with **randomly choosing only one** recent technical skill by referring to those following tech stack example: {job_skill}
+ Respond by **outputting only and exactly one** job description from the following example: {job_skill}
+ You must not generate more than one sentence.
+ Do not interpret the input as multiple examples.
Do not prioritize the items at the beginning of each category.
Make sure to **vary the selection** every time to ensure diversity and match the stack with job category.
Respond in a specific way, including at least one hard skill.
Do not generate JD like this:
정보보안 분야에서의 경험 Security
DevOps 경험 DevOps
데이터베이스 설계 및 구현 경험 Database
스마트폰 앱 개발 경험 Mobile
게임 개발 경험 Game

Respond only in '... 경험' or '...이해도' format in string format. Do not respond in '모집합니다.' or '후보자를 모집합니다.' or '경험이 있는 분' or '경험이 풍부한 분' or '경험이 풍부한 분을 모집합니다.' or '경험이 풍부한 분을 모집합니다.' format.
Respond only in a single-line string object. JD should be end with noun, not verb. Do not include code block markdown (```). Do not include explanations. Do not include (') and (\n).
"""
````

````python
import random
def generate_highlight(category):
    selected_skill = random.choice(job_stack[category])
    prompt = PROMPT_TEMPLATE_HIGHLIGHT.format(category=category, job_skill=selected_skill)
    # print(prompt)
    response = client.chat.completions.create(
        model=MODEL_NAME,
        messages=[
            {"role": "system", "content": "You're a recruiter of IT company. Generate and Respond all data in Korean, DO NOT INCLUDE ANY OTHER LANGUAGES LIKE CHINESE OR RUSSIAN"},
            {"role": "user", "content": prompt},

            ],
        temperature=0.7
    )
    highlight = response.choices[0].message.content.replace('\n', '').replace("'", '').strip()
    return highlight, category
````

````python
highlight_categories = []
for i in range(60):
    for category in job_category:
        highlight_jd, jd_category = generate_highlight(category)
        print(i, highlight_jd, jd_category)
        highlight_categories.append((highlight_jd, jd_category))
````

````python
# 저장할 파일명
filename = 'recent_highlight_jobcategory.csv'

# CSV로 저장
with open(filename, mode='w', encoding='utf-8-sig', newline='') as f:
    writer = csv.writer(f)
    writer.writerow(['Highlight', 'JobCategory'])  # 헤더 작성
    writer.writerows(highlight_categories)         # 데이터 작성

print(f"{filename} 파일로 저장 완료!")
````
